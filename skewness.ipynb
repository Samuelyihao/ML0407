```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
dist= np.random.normal(loc=10,size=100000)
dist = dist[dist >= 0]
sns.distplot(dist)
```

```python
from scipy.stats import boxcox_normmax
dist1 = np.power(dist, 6)
print(boxcox_normmax(dist1))
sns.distplot(dist1)
```

```python
dist1 = np.power(dist, 1/9)
print(boxcox_normmax(dist1))
sns.distplot(dist1)
```

```python
dist1 = np.power(2, dist)
print(boxcox_normmax(dist1))
sns.distplot(dist1)
```
